```python
# logical opertors(x is y)and (x is not y)
x = ["applr","banana"]
y = ["apple","banana"]
z = x
print(x is z)

# returns true because z is the same object as x

print(x is y)

# returns false because x is not the same object as y,even if they have the same content

print(x == y)
# to demonstrate the difference between "is " and "==" this  comparison returns true because x is equal toy



x = ["applr","banana"]
y = ["apple","banana"]
z = x
print(x is not z)

# returns true because z is the same object as x

print(x is not y)

# returns false because x is not the same object as y,even if they have the same content

print(x != y)
# to demonstrate the difference between "is " and "==" this  comparison returns true because x is equal to y















```

```python
# membership opertors (x in y) and (x not in y)

x = ["apple","banana"]

print("banana" in x )

# returns true because a sequence with the value "banana" is in the list



x = ["apple","banana"]

print("pineapple" not in x )

# returns true because a sequence with the value "pineapple" is not in the list


x = ["apple","banana"]

print("pineapple" in x )

# returns true because a sequence with the value not "pineapple" is in the list



x = ["apple","banana"]

print("banana" not in x )

# returns true because a sequence with the value "banana" is  in the list





```

```python
# bitwise opertors (x and y),(x / y),(x^y),(~x),(x<<2),(x>>2)

x = 3
y = 6
print(x & y)      #(and)

print(5 & 8)      #(and)


x = 3
y = 6
print(x | y)      #(or)

print(5 | 8)      #(or)



x = 3
y = 6
print(x ^ y)      #(xor)

print(5 ^ 8)      #(xor)


x = 3
y = 6
print( ~ y)      #(not)

print( ~ 8)      #(not)

x = 3
y = 6
print(x << y)      #(zero fill left shift)

print(5 << 8)      #(zero fill left shift)


x = 3
y = 6
print(x >> y)      #(signed right shift)

print(5 >> 8)      #(zero fill left shift)











```
